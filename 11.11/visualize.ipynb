```python
import pandas as pd

from sklearn.model_selection import train_test_split

df = pd.read_csv('/content/preprocessing.csv')

df.head()
```

```python
df_1 = df[['Rent day','total count']]

df_1.head()

```

```python
import matplotlib.pyplot as plt
import numpy as np

#크기
plt.figure(figsize=(30,8))

#x축 레이블 기울기 설정
plt.xticks(rotation='45')

#그래프
plt.plot(df_1.groupby('Rent day').sum()['total count']) # 하루 총 대여 수
#plt.plot(df_1.groupby('Rent day').mean()['total count']) # 하루 평균 대여 수

#x축 값 간격 설정
plt.xticks([0,30,60,90,120,150,180,210,240,270,300,330], labels=['2020-07-01', '2020-08-01', '2020-09-01', '2020-10-01', '2020-11-01', '2020-12-01', '2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01', '2021-05-01'])

#x, y축 레이블 값
plt.xlabel('date', size = 20)
plt.ylabel('num', size=20)

plt.show()
```
