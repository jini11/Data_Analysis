```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

data = requests.get('https://music.bugs.co.kr/chart')
soup = BeautifulSoup(data.text, 'html.parser')

ranking_list = []
m_title_list = []
m_artist_list = []
m_album_list = []

rank_lists = soup.select('.list > tbody > tr')


for rank_list in rank_lists:
    ranking = rank_list.select_one('td > div.ranking > strong').text
    ranking_list += [ranking]
    m_title = rank_list.select_one('th > p.title > a').text
    m_title_list += [m_title]
    m_artist = rank_list.select_one('td.left > .artist > a').text
    m_artist_list += [m_artist]
    m_album = rank_list.select_one('td.left > a ').text
    m_album_list += [m_album]
    print(ranking + '    ' + m_title + '    ' + m_artist + '    ' + m_album)

music_chart_df = pd.DataFrame({"Rank" : ranking_list,
                               "Title" : m_title_list,
                              "Artist" : m_artist_list,
                               "Album" : m_album_list})

music_chart_df.to_csv ('/content/sample_data/music_chart_df.csv')

music_chart_df
```
