```python
import requests
import uuid
import time
import json
import pprint

api_url = 'https://668ac839895c4ab0a6770b88e769f030.apigw.ntruss.com/custom/v1/12528/5f308e1810ab7cf66fc39768fa80810b83da70af413b09a44d082d0063bdc5be/general'
secret_key = ''
#image_file = '/content/Screenshot_20210819-203622_Starbucks.jpg'
image_file = '/content/등기부등본.png'

request_json = {
    'images': [
        {
            'format': 'jpg',
            'name': 'demo'
        }
    ],
    'requestId': str(uuid.uuid4()),
    'version': 'V2',
    'timestamp': int(round(time.time() * 1000)),
    'lang': 'ko'
}

payload = {'message': json.dumps(request_json).encode('UTF-8')}
files = [
  ('file', open(image_file,'rb'))
]
headers = {
  'X-OCR-SECRET': secret_key
}

response = requests.request("POST", api_url, headers=headers, data = payload, files = files)


json_data = json.loads(response.text)
pprint.pprint(json_data)
```
